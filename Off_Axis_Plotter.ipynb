# Justin's Tools for calculating off-axis companion glint
This tool uses the unocculted HLC PSF profile from: https://roman.ipac.caltech.edu/data/sims/coronagraph/perf_files/off-axis.pdf to understand the brightness of off-axis companions to reference/science primaries and how much flux they introduce into the dark hole (in 1D).

This is all captured in OffAxisPlotter.py, which has a built-in function to create a plot for you based off an input primary star magnitude, companion magnitude, and companion separation. For this to work, the file "HLC_prof.csv" must be in your current working directory. Easiest to clone the repository: https://github.com/jrhom1/CorGI_ObsTools/tree/main

```python
import OffAxisPlotter as OffAP
```

```python
#Magnitudes are V magnitudes. Separation is in arcseconds.
OffAP.offaxisPSF_Demo(primarymag=2.5,comp_sep=1.0,comp_mag=10,save=False)
```
